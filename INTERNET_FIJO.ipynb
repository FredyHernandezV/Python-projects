#MODIFICACIONES A LA BASE DE DATOS
Debido a que se realizara un enfoque hacia el departamento de Cundinamarca y Bogota, es necesario realizar un filtro, el cual brinde un nuevo DataFrame, que solo contenga los datos pertenecientes a dicho municipio y ciudad.



##MODIFICACION DE TILDES EN BASE DE DATOS
Se realiz√≥ un ajuste en toda la base de datos para eliminar las tildes de las palabras.

Esto se hizo porque exist√≠an registros duplicados de un mismo municipio: algunos estaban guardados con tilde y otros sin ella, lo que generaba inconsistencias en la informaci√≥n.

```python
df_InternertFijo_MOD = df_InternertFijo.applymap(
    lambda x: unidecode(x) if isinstance(x, str) else x
)
df_InternertFijo_MOD
```

##FILTRANDO BASE DE DATOS PARA USAR UNICAMENTE DATOS DE CUNDINAMARCA Y BOGOT√Å

```python
df_InternertFijo_Cun = df_InternertFijo_MOD[(df_InternertFijo_MOD['DEPARTAMENTO']== "CUNDINAMARCA") | (df_InternertFijo_MOD['DEPARTAMENTO']== "BOGOTA D.C.")]
df_InternertFijo_Cun
```

##VERIFICACION DE DATOS

```python

duplicados_total = df_InternertFijo_Cun.duplicated().sum()
print(f"Cantidad de filas duplicadas: {duplicados_total}")


if duplicados_total > 0:
    print("\nFilas duplicadas:")
    display(df_InternertFijo_Cun[df_InternertFijo_Cun.duplicated(keep=False)])
else:
    print("No existen filas duplicadas en el dataset.")
```

```python
# Eliminar filas duplicadas y conservar solo la primera aparici√≥n
df_InternertFijo_Cun = df_InternertFijo_Cun.drop_duplicates(keep='first')

# Verificar nuevamente que no existan duplicados
print(f"Cantidad de filas duplicadas despu√©s de eliminar: {df_InternertFijo_Cun.duplicated().sum()}")

# Confirmar tama√±o final del DataFrame
print(f"N√∫mero total de filas despu√©s de limpiar: {len(df_InternertFijo_Cun)}")
```

##AJUSTE TIPO DE VARIABLES

```python
df_InternertFijo_Cun["COD_DEPARTAMENTO"] = df_InternertFijo_Cun["COD_DEPARTAMENTO"].astype(str)
df_InternertFijo_Cun["COD_MUNICIPIO"] = df_InternertFijo_Cun["COD_MUNICIPIO"].astype(str)

# Limpiar y convertir velocidades (por si vienen con comas o textos)
df_InternertFijo_Cun["VELOCIDAD_BAJADA"] = (
    df_InternertFijo_Cun["VELOCIDAD_BAJADA"].astype(str)
    .str.replace(",", ".")
    .str.extract(r"(\d+\.?\d*)")[0]
    .astype(float)
)

df_InternertFijo_Cun["VELOCIDAD_SUBIDA"] = (
    df_InternertFijo_Cun["VELOCIDAD_SUBIDA"].astype(str)
    .str.replace(",", ".")
    .str.extract(r"(\d+\.?\d*)")[0]
    .astype(float)
)

df_InternertFijo_Cun.info()
```

#KPI Y METRICAS

```python
# KPI 1: Total de Accesos Fijos
total_accesos = df_InternertFijo_Cun['No DE ACCESOS'].sum()

print("--- 1. Total Global de Accesos Fijos ---")
# Usamos formato de miles para mejor lectura
print(f"Total de Accesos Fijos: {total_accesos:,.0f}")
```

```python
# KPI 2: Crecimiento Anual de Accesos

# 1. Agrupar la suma de accesos por a√±o
accesos_por_a√±o = df_InternertFijo_Cun.groupby('A√ëO')['No DE ACCESOS'].sum().reset_index()
accesos_por_a√±o.columns = ['A√ëO', 'Accesos']

# 2. Calcular el crecimiento porcentual (pct_change)
# Se usa un "periodo" de 1 para comparar con la fila anterior (a√±o anterior)
accesos_por_a√±o['Crecimiento Anual (%)'] = accesos_por_a√±o['Accesos'].pct_change() * 100

print("\n--- 2. Crecimiento Anual de Accesos ---")
# El primer a√±o siempre ser√° NaN porque no hay dato previo para comparar.
accesos_por_a√±o
```

```python
# KPI 3: Participaci√≥n Geogr√°fica

# 1. Calcular el total de accesos por municipio
accesos_por_municipio = df_InternertFijo_Cun.groupby('MUNICIPIO')['No DE ACCESOS'].sum().reset_index()
accesos_por_municipio.columns = ['MUNICIPIO', 'Accesos']

# 2. Obtener el total global de accesos (reutilizamos el valor del KPI 1)
total_global_accesos = accesos_por_municipio['Accesos'].sum()

# 3. Calcular la participaci√≥n porcentual
accesos_por_municipio['Participaci√≥n Geogr√°fica (%)'] = (
    accesos_por_municipio['Accesos'] / total_global_accesos
) * 100

# 4. Mostrar el Top 10 de municipios con mayor participaci√≥n
participacion_top_10 = accesos_por_municipio.sort_values(
    by='Participaci√≥n Geogr√°fica (%)', ascending=False
).head(10)

print("\n--- 3. Participaci√≥n Geogr√°fica (Top 10 Municipios) ---")
participacion_top_10
```

```python
# KPI 4: Velocidad Mediana de Bajada (VMB)

# A. VMB Global
vmb_global = df_InternertFijo_Cun['VELOCIDAD_BAJADA'].median()
print("--- 4. Velocidad Mediana de Bajada (VMB) ---")
print(f"VMB Global (P50): {vmb_global:.2f} Mbps")
print("-" * 30)

# B. VMB por Municipio (Para comparar la calidad geogr√°fica)
vmb_por_municipio = df_InternertFijo_Cun.groupby('MUNICIPIO')['VELOCIDAD_BAJADA'].median().reset_index()
vmb_por_municipio.columns = ['MUNICIPIO', 'VMB (Mbps)']

print("\nTop 10 VMB por Municipio:")
vmb_por_municipio.sort_values(by='VMB (Mbps)', ascending=False).head(10)
```

```python
# KPI 5: Velocidad Promedio por Tecnolog√≠a

# Calcular la media (promedio) y la mediana de la velocidad por cada tecnolog√≠a
velocidad_por_tecnologia = df_InternertFijo_Cun.groupby('TECNOLOGIA')['VELOCIDAD_BAJADA'].agg(
    Media_Mbps = 'mean',
    Mediana_Mbps = 'median'
).reset_index()

# Ordenar por Media de Bajada para ver el ranking de desempe√±o
velocidad_por_tecnologia = velocidad_por_tecnologia.sort_values(by='Media_Mbps', ascending=False)

print("\n--- 5. Velocidad Promedio (Media y Mediana) por Tecnolog√≠a ---")
velocidad_por_tecnologia
```
