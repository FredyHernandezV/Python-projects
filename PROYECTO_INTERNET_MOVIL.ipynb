#¿De que Trata la Base de Datos?

La base de datos “Internet Fijo – Accesos por tecnología y segmento” contiene información sobre el número de suscriptores con acceso dedicado a Internet fijo en Colombia, reportado por los diferentes proveedores de servicios al Ministerio de Tecnologías de la Información y las Comunicaciones (MinTIC).

Los datos se presentan de manera agregada (no individual), y corresponden al número total de conexiones activas por departamento, municipio, proveedor, tecnología, segmento y trimestre.

```python

# 1. Agrupar los datos para el análisis bivariado: Suma de accesos por Municipio y Tecnología.
# Esto nos dice cuántos accesos tiene cada tecnología en cada municipio.
accesos_tecnologia_municipio = df_InternertFijo_Cun.groupby(['MUNICIPIO', 'TECNOLOGIA'])['No DE ACCESOS'].sum().reset_index()

# 2. Ordenar los municipios para que Bogotá D.C. aparezca primero o usar solo un Top N
# Opcional: Si tienes demasiados municipios, puedes filtrar para mostrar solo los más relevantes
# Por ejemplo, solo los 5 municipios con más accesos + Bogotá.
top_municipios = df_InternertFijo_Cun.groupby('MUNICIPIO')['No DE ACCESOS'].sum().nlargest(5).index.tolist()
if 'BOGOTA D.C.' not in top_municipios and 'BOGOTA D.C.' in df_InternertFijo_Cun['MUNICIPIO'].unique():
    top_municipios.append('BOGOTA D.C.')
    top_municipios = list(set(top_municipios)) # Elimina duplicados

accesos_filtrados = accesos_tecnologia_municipio[accesos_tecnologia_municipio['MUNICIPIO'].isin(top_municipios)]


# 3. Creación del Gráfico de Barras Agrupadas
fig_tec_mun = px.bar(
    accesos_filtrados,
    x='MUNICIPIO',
    y='No DE ACCESOS',
    color='TECNOLOGIA', # Usamos TECNOLOGIA para el color
    barmode='stack',   # 'group' o 'stack'. 'stack' muestra la participación total del municipio, 'group' facilita la comparación entre tecnologías.
    title='🌐 Distribución y Participación de Tecnologías por Municipio (Top)',
    color_discrete_sequence=px.colors.qualitative.Prism,
    hover_data={'No DE ACCESOS': ':.0f'} # Formato de miles en el hover
)

fig_tec_mun.update_layout(
    yaxis_title='Número de Accesos',
    xaxis_title='Municipio',
    legend_title_text='Tecnología',
    xaxis={'categoryorder': 'total descending'} # Ordena los municipios por el total de accesos
)
fig_tec_mun.show()
```

#ANALISIS UNIVARIADO

```python
# --- 2. PREPARACIÓN DE TIPOS DE DATOS ---
df_InternertFijo_Cun['No DE ACCESOS'] = pd.to_numeric(df_InternertFijo_Cun['No DE ACCESOS'], errors='coerce')
df_InternertFijo_Cun['VELOCIDAD_BAJADA'] = pd.to_numeric(df_InternertFijo_Cun['VELOCIDAD_BAJADA'], errors='coerce')
df_InternertFijo_Cun['VELOCIDAD_SUBIDA'] = pd.to_numeric(df_InternertFijo_Cun['VELOCIDAD_SUBIDA'], errors='coerce')
df_InternertFijo_Cun.dropna(subset=['No DE ACCESOS', 'VELOCIDAD_BAJADA', 'VELOCIDAD_SUBIDA'], inplace=True)


# ======================================================================
## 1. Análisis Univariado de Evolución Temporal (Conteo de Registros por Año) - OE 1
# ======================================================================

print("--- 1. Evolución Temporal (Conteo de Registros) ---")
conteo_anual = df_InternertFijo_Cun['AÑO'].value_counts().sort_index().reset_index()
conteo_anual.columns = ['AÑO', 'Cantidad']

fig = px.bar(
    conteo_anual,
    x='AÑO',
    y='Cantidad',
    text='Cantidad',
    title='📊 Conteo de Registros por Año (2017-2023)',
    color='Cantidad',
    color_continuous_scale='YlGnBu'
)
fig.update_traces(textposition='outside')
fig.update_layout(xaxis_title='Año', yaxis_title='Número de Registros')
fig.show()
print(conteo_anual)

# --- Análisis por Total de Accesos por Año (Medida más robusta) ---
print("\n--- Total de Accesos Sumados por Año ---")
accesos_anual = df_InternertFijo_Cun.groupby('AÑO')['No DE ACCESOS'].sum().reset_index()

fig_acc = px.line(
    accesos_anual,
    x='AÑO',
    y='No DE ACCESOS',
    markers=True,
    title='📈 Evolución Anual del Total de Accesos (2017-2023)',
)
fig_acc.update_layout(yaxis_title='Total de Accesos (Suma)', xaxis_title='Año')
fig_acc.show()
print(accesos_anual)

# ======================================================================
## 2. Análisis Univariado de Calidad (Velocidad) - OE 4
# ======================================================================

print("\n--- 2. Estadísticas Descriptivas de Velocidades (Calidad General) ---")
desc_velocidad = df_InternertFijo_Cun[['VELOCIDAD_BAJADA', 'VELOCIDAD_SUBIDA']].describe().transpose()
print(desc_velocidad)

# Visualización de la distribución de la Velocidad de Bajada (con Box Plot para atípicos)
fig_box = px.box(
    df_InternertFijo_Cun,
    y='VELOCIDAD_BAJADA',
    title='📦 Distribución de la Velocidad de Bajada (Detección de Atípicos)',
    points='all', # Muestra todos los puntos
    color_discrete_sequence=['#FF7F0E']
)
fig_box.show()

# ======================================================================
## 3. Análisis Univariado de Tecnología (Participación) - OE 3
# ======================================================================

print("\n--- 3. Participación por Tecnología (OE 3) ---")
tecnologia_accesos = df_InternertFijo_Cun.groupby('TECNOLOGIA')['No DE ACCESOS'].sum().reset_index()
tecnologia_accesos['Porcentaje'] = (tecnologia_accesos['No DE ACCESOS'] / tecnologia_accesos['No DE ACCESOS'].sum()) * 100
tecnologia_accesos = tecnologia_accesos.sort_values(by='No DE ACCESOS', ascending=False)

fig_tec = px.bar(
    tecnologia_accesos.head(10), # 👈 Grafica solo las 10 primeras filas
    x='TECNOLOGIA',
    y='No DE ACCESOS',
    # --- LÍNEA CORREGIDA ---
    # Aplica .head(10) a la columna de texto para que coincida con las barras.
    text=tecnologia_accesos['Porcentaje'].head(10).round(1).astype(str) + '%',
    # -----------------------
    title='🌐 Participación de Tecnologías de Conexión (Top 10 por Accesos)',
    color='No DE ACCESOS',
    color_continuous_scale='Viridis'
)
fig_tec.update_traces(textposition='outside')
fig_tec.update_layout(xaxis_title='Tecnología', yaxis_title='Total de Accesos (Suma)')
fig_tec.show()
print(tecnologia_accesos)

# ======================================================================
## 4. Análisis Univariado de Segmento (Socioeconómico)
# ======================================================================

print("\n--- 4. Distribución por Segmento Socioeconómico ---")
segmento_accesos = df_InternertFijo_Cun.groupby('SEGMENTO')['No DE ACCESOS'].sum().reset_index()
segmento_accesos = segmento_accesos.sort_values(by='No DE ACCESOS', ascending=False)

fig_seg = px.bar(
    segmento_accesos.head(10),
    x='SEGMENTO',
    y='No DE ACCESOS',
    title='🧑‍💻 Distribución de Accesos por Segmento (Socioeconómico y Corporativo)',
    color='No DE ACCESOS',
    color_continuous_scale='Plasma'
)
fig_seg.update_layout(xaxis_title='Segmento', yaxis_title='Total de Accesos (Suma)')
fig_seg.show()
print(segmento_accesos)
```

```python
import plotly.express as px

# Agrupar los accesos totales por tecnología
tecnologia_total = (
    df_InternertFijo_Cun.groupby('TECNOLOGIA')['No DE ACCESOS']
    .sum()
    .reset_index()
    .sort_values(by='No DE ACCESOS', ascending=False)
)

# Crear gráfico de pastel
fig = px.pie(
    tecnologia_total,
    values='No DE ACCESOS',
    names='TECNOLOGIA',
    title='Distribución porcentual de accesos por tecnología',
    color_discrete_sequence=px.colors.qualitative.Pastel,
    hole=0.3  # Si quieres estilo “dona”, cámbialo a 0.3 o elimínalo para pastel completo
)

# Personalizar etiquetas y estilo
fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    pull=[0.05]*len(tecnologia_total)  # para dar un pequeño efecto de separación
)

fig.update_layout(
    showlegend=True,
    legend_title_text='Tecnología',
    legend=dict(orientation='h', y=-0.2, x=0.3)
)

fig.show()
```

```python
import plotly.express as px

# Agrupar por año y tecnología
tecnologia_año = (
    df_InternertFijo_Cun
    .groupby(['AÑO', 'TECNOLOGIA'])['No DE ACCESOS']
    .sum()
    .reset_index()
)

# Lista de años únicos
años = sorted(tecnologia_año['AÑO'].unique())

# Crear figura base con el primer año
año_inicial = años[0]
df_inicial = tecnologia_año[tecnologia_año['AÑO'] == año_inicial]

fig = px.pie(
    df_inicial,
    values='No DE ACCESOS',
    names='TECNOLOGIA',
    title=f'Distribución porcentual de tecnologías en {año_inicial}',
    color_discrete_sequence=px.colors.qualitative.Set3,
    hole=0.3
)

# Crear frames manualmente (uno por cada año)
frames = []
for año in años:
    df_year = tecnologia_año[tecnologia_año['AÑO'] == año]
    frames.append(dict(
        method='update',
        label=str(año),
        args=[
            {'values': [df_year['No DE ACCESOS']],
             'labels': [df_year['TECNOLOGIA']]},
            {'title': f'Distribución porcentual de tecnologías en {año}'}
        ]
    ))

# Agregar menú interactivo (dropdown)
fig.update_layout(
    updatemenus=[{
        "buttons": frames,
        "direction": "down",
        "showactive": True,
        "x": 0.5,
        "xanchor": "center",
        "y": -0.2,
        "yanchor": "top"
    }],
    legend_title_text='Tecnología',
    title_x=0.5
)

fig.show()
```
